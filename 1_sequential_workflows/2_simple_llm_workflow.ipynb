```python
from langgraph.graph import StateGraph, START, END
from langchain_groq import ChatGroq
from typing import TypedDict
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
GROQ_API_KEY=os.getenv("GROQ_API_KEY")
GROQ_MODEL=os.getenv("GROQ_MODEL")
```

```python
model = ChatGroq(model=GROQ_MODEL, api_key=GROQ_API_KEY)
```

```python
# define state
class LLMState(TypedDict):
    question: str
    answer: str
```

```python
def llm_qa(state: LLMState) -> LLMState:
    # extract question from state
    question = state['question']
    
    # form a prompt
    prompt = f"I would like you to answer the following question\n{question}"
    
    # ask that question to llm
    answer = model.invoke(prompt).content
    
    # update the answer in the state
    state['answer'] = answer
    
    return state
```

```python
# define graph
graph = StateGraph(LLMState)

# add node to the graph
graph.add_node('llm_qa', llm_qa)

# add edges to the node
graph.add_edge(START, 'llm_qa')
graph.add_edge('llm_qa', END)

# compile the graph
workflow = graph.compile()
```

```python
# execute 
initial_state = {'question': 'What are the distance from earth to moon?'}

final_state = workflow.invoke(initial_state)

print(final_state['answer'])
```
